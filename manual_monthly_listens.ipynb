```python
def get_value_for_artist(wanted_act, key, array):
    for artist in array:
        if artist['name'] == wanted_act:
            return artist[key]
    return 0
```

```python
import spotipy
import json
from spotipy.oauth2 import SpotifyClientCredentials
```

```python
def get_client_credentials(file_path="creds.json"):
    try:
        with open(file_path, 'r') as file:
            data = json.load(file)
            client_id = data['client_id']
            client_secret = data['client_secret']
            return client_id, client_secret
    except (KeyError, FileNotFoundError, json.decoder.JSONDecodeError) as e:
        print(f"No Spotify credentials found. {e}. Setting all acts to a popularity of 0.")
        return None, None
```

```python
# Authenticate with Spotify
client_id, client_secret = get_client_credentials()
auth_manager = SpotifyClientCredentials(client_id=client_id, client_secret=client_secret)
sp = spotipy.Spotify(auth_manager=auth_manager)
```

```python
#  {'name': '2Hollis', 'followers': 236642, 'popularity': 65}
best = []
for act in manualSpotifyScrape:
    id = act['id']
    if id == '0':
        monthly_listeners = 0
        artist_info = 0
        followers = 0
        popularity = 0
        name = act['name']
    else:
        monthly_listeners = act['monthly_listeners']
        artist_info = sp.artist(act['id'])
        followers = artist_info['followers']['total']
        popularity = artist_info['popularity']
        name = artist_info['name']
    act_info = {'name': name, 'followers': followers, 'popularity': popularity, "monthly_listeners" : monthly_listeners}
    best.append(act_info)
```

```python
duoActs = {"Torren Foot B2B Airwolf Paradise" : ["Torren Foot", "Airwolf Paradise"]}
duo_mult = 1.2  # Since duos are more hype, add a multiplier to their popularity and follower averages
for duo, artists in duoActs.items():
    followers = popularity = monthly_listeners = 0
    for artist in artists:
        popularity += get_value_for_artist(artist, 'popularity', best)
        followers += get_value_for_artist(artist, 'followers', best)
        monthly_listeners += get_value_for_artist(artist, 'monthly_listeners', best)
    followers = int((followers / len(artists)) * duo_mult)
    popularity = int((popularity / len(artists)) * duo_mult)
    monthly_listeners = int((monthly_listeners / len(artists)) * duo_mult)
    print([followers, popularity, monthly_listeners])
```
